```python
import pandas as pd
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix

from sklearn.svm import SVC
```

```python
df = pd.read_csv("emails.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df['Prediction']
```

```python
X = df.drop('Prediction'  , axis = 'columns').drop('Email No.' , axis='columns')
```

```python
X
```

```python
df
```

```python
y = df['Prediction']
```

```python
from sklearn.neighbors import KNeighborsClassifier
#import sklearn as sn
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test , y_train ,y_test = train_test_split(X,y, test_size=0.20 ,random_state=50)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=2)
```

```python
model_knn.fit(X_train , y_train)
```

```python
y_pred_knn = model_knn.predict(X_test);
```

```python
y_pred_knn
```

```python
print("Accuracy for KNN")
accuracy_score(y_pred_knn , y_test)
```

```python
print("Mean Square Error for KNN")
mean_squared_error(y_pred_knn , y_test)
```

```python
print(" Root Mean Square Error for KNN")
mean_squared_error(y_pred_knn , y_test , squared=False)
```

```python
confusion_matrix(y_pred_knn , y_test)
```

```python
model_svm = SVC()
```

```python
model_svm.fit(X_train , y_train)
```

```python
y_pred_svm = model_svm.predict(X_test)
```

```python
print("Accuracy for SVM")
accuracy_score(y_pred_svm , y_test)
```

```python
print("Mean Square Error for SVM")
mean_squared_error(y_pred_svm , y_test)
```

```python
print("Root Mean Square Error for SVM")
mean_squared_error(y_pred_svm , y_test , squared=False)
```

```python
confusion_matrix(y_pred_svm , y_test)
```
